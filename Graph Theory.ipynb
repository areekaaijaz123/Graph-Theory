```python
import networkx as nx
import matplotlib.pyplot as plt

G=nx.Graph()
G.add_node("a")
G.add_node("b")
G.add_node("c")

G.add_edge("a","b",weight=100)
G.add_edge("a","c",weight=80)
G.add_edge("e","a",weight=70)
G.add_edge("d","b",weight=60)
G.add_edge("d","e",weight=50)
G.add_edge("c","e",weight=40)

pos=nx.circular_layout(G)
nx.draw(G,pos,with_labels=1)
plt.show()
```

```python
import networkx as nx
import matplotlib.pyplot as plt

G = nx.DiGraph()

print("Choose any input file")
print(" 1. input10")
print(" 2. input20")
print(" 3. input30")
print(" 4. input40")
print(" 5. input50")
print(" 6. input60")
print(" 7. input70")
print(" 8. input80")
print(" 9. input90")
print(" 10. input100")
file = input("Enter file : ")

if file == "1":
    f = open("input10.txt","r") 
    nodes = 10
elif file == "2":
    f = open("input20.txt","r") 
    nodes = 20
elif file == "3":
    f = open("input30.txt","r") 
    nodes = 30
elif file == "4":
    f = open("input40.txt","r") 
    nodes = 40
elif file == "5":
    f = open("input50.txt","r")
    nodes = 50
elif file == "6":
    f = open("input60.txt","r")
    nodes = 60
elif file == "7":
    f = open("input70.txt","r")
    nodes = 70
elif file == "8":
    f = open("input80.txt","r")
    nodes = 80
elif file == "9":
    f = open("input90.txt","r")
    nodes = 90
elif file == "10":
    f = open("input100.txt","r")
    nodes = 100

for i in range(0,nodes+5):
    f.readline()
     
for line in f:
    arr_list = []
    for word in line.split():
        arr_list.append(word)    
    if not line.strip():
        break   
    for x in range(1,len(arr_list),4):
        w = arr_list[x+2]
        if G.has_edge(arr_list[0],arr_list[x])==0:
            G.add_edge(arr_list[0],arr_list[x],weight=w)
            
pos=nx.circular_layout(G)
nx.draw(G,pos,with_labels=1)
plt.show()

start = f.readline()
print(start)
```

```python
print("Prim's Algorithm")

Prim = nx.Graph() 

visited=['1']

while (len(Prim) < len(G)):
    min = '99999999999999'
    i='0'
    j='0'
    for u in G.node():
        if u in visited:
            for v in G.node():
                if v not in visited and G.has_edge(u,v)==1 :
                    if min > G[u][v]['weight']:
                        min = G[u][v]['weight']
                        i = u
                        j = v
    visited.append(j)
    Prim.add_edge(i,j)                 
        
pos=nx.circular_layout(Prim)
nx.draw(Prim,pos,with_labels=1)
plt.show()       
```
